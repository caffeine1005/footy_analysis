```python
epl = pd.read_csv('prem_2025.csv')

home_manutd = epl[epl['home_team'] == 'Manchester United']
away_manutd = epl[epl['away_team'] == 'Manchester United']

manutd = pd.concat([home_manutd, away_manutd])

manutd
```

```python
list = manutd['game_id'].to_list()
for i in range(len(list)):
    print(list[i])
```

```python
utd_game_list = []

for i in range(len(list)):
    game = ws.read_events(match_id = list[i])
    utd_game_list.append(game)
```
